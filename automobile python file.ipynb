# IQR Method

```python
column_name = 'engine-size'

# Calculate the IQR (Interquartile Range)
Q1 = df_automobile[column_name].quantile(0.25)
Q3 = df_automobile[column_name].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df_automobile[(df_automobile[column_name] < lower_bound) | (df_automobile[column_name] > upper_bound)]

# Print the outliers
print("Outliers:")
print(outliers)
```

```python
# Define a function to find outliers in a dataset using the IQR method
def find_outliers_iqr(data=df_automobile):
    # Calculate the first quartile (Q1) and third quartile (Q3)
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)

    # Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1

    # Calculate the lower and upper bounds to identify outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Return a boolean mask indicating which data points are outliers
    return (data < lower_bound) | (data > upper_bound)
outliers = df_automobile[['engine-size']].apply(find_outliers_iqr)

# Display rows containing potential outliers
potential_outliers = df_automobile[outliers.any(axis=1)]
print("Rows with Potential Outliers:")
potential_outliers
```

# Hypothesis Testing


### Two-Sample T-test Hypothesis Test

```python
# Subset the dataset for diesel and gasoline cars
diesel_prices = df_automobile[df_automobile['fuel-type'] == 'diesel']['price']
gasoline_prices = df_automobile[df_automobile['fuel-type'] == 'gasoline']['price']

# Perform a two-sample t-test
t_statistic, p_value = stats.ttest_ind(diesel_prices, gasoline_prices, equal_var=False)

# Set the significance level (alpha)
alpha = 0.05

# Print the results
print(f'Test Statistic: {t_statistic}')
print(f'P-value: {p_value}')

# Compare p-value to the significance level
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

Null Hypothesis (H0): The average price of diesel-fueled cars is equal to the average price of gasoline-fueled cars.
Alternative Hypothesis (H1): The average price of diesel-fueled cars is different from the average price of gasoline-fueled cars.
The ttest_ind function from scipy.stats is used for the two-sample t-test. The equal_var=False parameter indicates that we assume unequal variances between the two groups.

### Chi-Squared () Hypothesis Test

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df_automobile['fuel-type'], df_automobile['body-style'])

# Perform the chi-squared test
chi2_stat, p_value, _, _ = chi2_contingency(contingency_table)

# Set the significance level (alpha)
alpha = 0.05

# Print the results
print(f'Chi-squared Statistic: {chi2_stat}')
print(f'P-value: {p_value}')

# Compare p-value to the significance level
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

# Distribution Analysis

```python
from scipy.stats import shapiro


# Select a column for distribution analysis ('price')
column_name = 'price'
data = df_automobile[column_name].dropna()

# Plot a histogram and kernel density estimate (KDE) for visual inspection
sns.histplot(data, kde=True)
plt.title(f'Distribution of {column_name}')
plt.show()

# Perform Shapiro-Wilk test for normality
stat, p_value = shapiro(data)
print(f'Shapiro-Wilk Test - Statistic: {stat}, P-value: {p_value}')

# Interpret the results
if p_value < 0.05:
    print(f'The distribution of {column_name} is not normal.')
else:
    print(f'The distribution of {column_name} appears to be normal.')
```
